```python
import pandas as pd
import pandera as pa # Utilizado para validar data
```

```python
## Converter os dodos data para data
## ajustar a ordem da data: dayfirst=True
df = pd.read_csv("ocorrencia.csv", sep=";", parse_dates=['ocorrencia_dia'], dayfirst=True)
#df # imprime todas as linhas
df.head(10) # imprime as primerias 10 linhas

## Codigo Nat - Sem informacao de data campo data
## Codigo NaN - Campo vazio campo texto
```

```python
### Validando os dados em tempo real com schema
```

```python
schema = pa.DataFrameSchema( 
    columns = {
    #"codigo":pa.Column(pa.Int, required=False), add column that is not exist (required=False)
    "codigo_ocorrencia":pa.Column(pa.Int),
    "codigo_ocorrencia2":pa.Column(pa.Int),
    "ocorrencia_classificacao":pa.Column(pa.String),
    "ocorrencia_cidade":pa.Column(pa.String),
    "ocorrencia_uf":pa.Column(pa.String, pa.Check.str_length(2,2)),
    "ocorrencia_aerodromo":pa.Column(pa.String),
    "ocorrencia_dia":pa.Column(pa.DateTime),
    "ocorrencia_hora":pa.Column(pa.String, pa.Check.str_matches(r'ˆ([0-1]?[0-9]|[2][0-3]):([0-5][0-5][0-9])?$'), nullable=True), 
    "total_recomendacoes":pa.Column(pa.Int)
    
    }
)
```

```python
schema.validate(df)
```
